```python
good_weather = trips[trips['weather_conditions'] == 'Good']['duration_seconds']
bad_weather = trips[trips['weather_conditions'] == 'Bad']['duration_seconds']
```

```python
avg_good_weather = good_weather.mean()
avg_bad_weather = bad_weather.mean()
print(f"The average duration of good weather: {avg_good_weather}")
print(f"The average duration of bad weather: {avg_bad_weather}")
```

#### As the average is high for bad weather it looks like based on the data Chicago neighborhood the cabs take longer time during stormy and rainy days.

##### Null hypothesis (H0): The duration of rides on rainy day or any other day is same
##### Alternate hypothesis (H1): The duration of rides on rainy day or any other day is different.

```python
# Testing the hypothesis between good & bad weather
t_stat, p_value = ttest_ind(good_weather, bad_weather)

alpha= 0.05

print(f"T_statistics: {t_stat}")
print(f"P_value: {p_value}")

if (p_value < alpha):
    print ("we can reject the hypothesis")
else:
    print ("we can't reject the hypothesis")
    
```

##### Null hypothesis: The duration of rides on rainy day or non rainy day  is same. Alternate hypothesis the duration of rides on rainy day or non rainy day is different. As we are rejecting the hypothesis it means based on the analysis the weather duration between rainy day and non rainy days is different.  As p value is 7.36 so there is 700 % probability of the duration being different on both days.

```python
trips['start_ts'] = pd.to_datetime(trips['start_ts'])   
trips['dayofweek'] = trips['start_ts'].dt.day_name()
```

```python
no_rainy_sat = trips[(trips['weather_conditions'] == 'Good') & (trips['dayofweek'] == 'Saturday')]['duration_seconds']
rainy_sat = trips[(trips['weather_conditions'] =='Bad')& (trips['dayofweek'] == 'Saturday')]['duration_seconds']
```

##### Null hypothesis (H0): The duration of rides on rainy Saturday or any other Saturday is same
##### Alternate hypothesis (H1): The duration of rides on rainy Saturday or any other Saturday is different.

```python
## Testing hypothesis between rainy and non rainy saturday
t_stat, p_value = ttest_ind(no_rainy_sat, rainy_sat )

alpha= 0.05

print(f"T_statistics: {t_stat}")
print(f"P_value: {p_value}")

if (p_value < alpha):
    print ("we can reject the hypothesis")
else:
    print ("we can't reject the hypothesis")
    
```

##### Null hypothesis: The duration of rides on rainy Saturday or any other Saturday is same. Alternate hypothesis the duration of rides on rainy Saturday or any other Saturday is different. As we are rejecting the hypothesis it means based on the analysis that the duration between rainy Saturday and other Saturday is different. As p value is 7.36 so there is 700 % probability of the duration being different on both saturdays.
